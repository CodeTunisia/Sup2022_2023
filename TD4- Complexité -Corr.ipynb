# Exercice 5

```python
def puissance_naive(x, n):
    result = 1
    for i in range(n):
        result *= x
    return result
```

```python
print(puissance_naive(2, 10))  # affiche 1024
print(puissance_naive(3, 5))   # affiche 243
print(puissance_naive(5, 3))   # affiche 125
```

La complexité de cette fonction en temps d'exécution est en O(n), car elle effectue n multiplications. Cette complexité est plus grande que celle de l'algorithme d'exponentiation rapide, qui est en O(log n).

```python
def puissance2(x, n):
    if n == 0:
        return 1
    elif n % 2 == 0:
        return puissance2(x*x, n//2)
    else:
        return x * puissance2(x*x, n//2)
```

```python
print(puissance2(2, 10))  # affiche 1024
print(puissance2(3, 5))   # affiche 243
print(puissance2(5, 3))   # affiche 125
```

Cette fonction utilise la récursion et l'optimisation de l'algorithme d'exponentiation rapide consistant à réduire la valeur de n à chaque étape. Si n est pair, on peut écrire x^n comme (x^(n/2))^2. Cela signifie qu'on peut calculer x^n en calculant x^(n/2) et en le multipliant par lui-même. Si n est impair, on peut écrire x^n comme x * (x^((n-1)/2))^2. Cela signifie qu'on peut calculer x^n en multipliant x par le résultat de (x^((n-1)/2))^2.

En utilisant cette optimisation, on réduit la valeur de n à chaque étape de manière à réduire le nombre de calculs nécessaires. Par exemple, pour calculer x^10, on peut utiliser l'optimisation de l'algorithme pour réduire la valeur de n à chaque étape de manière à ne faire que trois multiplications au lieu de dix.

L'algorithme d'exponentiation rapide a une complexité en temps d'exécution en O(log n).

Pour démontrer cela, on peut utiliser un raisonnement en termes de nombre d'opérations effectuées pour démontrer que l'algorithme a une complexité en O(log n). À chaque étape de l'algorithme, on divise la valeur de n par 2, de sorte que le nombre d'étapes est logarithmique en n.

Voici comment écrire une version itérative de la fonction d'exponentiation rapide :


```python
def puissance2_iterative(x, n):
    result = 1
    while n > 0:
        if n % 2 == 1:
            result = result * x
        x = x * x
        n = n // 2
    return result
```

```python
print(puissance2_iterative(2, 10))  # affiche 1024
print(puissance2_iterative(3, 5))   # affiche 243
print(puissance2_iterative(5, 3))   # affiche 125
```

La complexité de cette fonction en temps d'exécution est en O(log n), comme pour la version récursive de l'algorithme.

# Exercice 6

```python
def liste_symetrique_avec_liste_intermediaire(L):
    # On crée une liste vide pour stocker les éléments de L dans leur ordre inversé
    L_inverse = []
    # On parcourt la liste L à l'envers en utilisant une boucle for
    for i in range(len(L)-1, -1, -1):
        # On ajoute chaque élément de L dans l'ordre inverse à L_inverse
        L_inverse.append(L[i])
    # On retourne la liste inverse
    return L_inverse

def liste_symetrique_sans_liste_intermediaire(L):
    # On parcourt la liste L en utilisant une boucle for
    for i in range(len(L)//2):
        # On échange chaque élément de L avec son symétrique
        L[i], L[-i-1] = L[-i-1], L[i]
    # On retourne la liste modifiée
    return L
```

Les deux fonctions sont de complexité en O(n). La deuxième fonction est généralement plus rapide que la première, car elle ne crée pas de liste intermédiaire et n'a pas besoin de parcourir la liste L à l'envers. Cependant, elle modifie la liste L en place, ce qui peut être un inconvénient dans certains cas. Si vous avez besoin de garder la liste d'origine intacte, vous devriez utiliser la première fonction qui crée une liste intermédiaire.
